# Existe uma Personalidade mais atrativa para a indústria cinematográfica?

O Objetivo dessa Análise é desvendar a influência de tipos específicos de personalidades na popularidade de filmes.
Nesse Notebook utilizarei dois conjuntos de dados criados por mim.




```python
#Importação de Bibliotécas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Importação dos Datasets
dt1 = pd.read_csv("https://raw.githubusercontent.com/Luanamero/FilmesMBTI/main/OscarMbti.csv")
dt2 = pd.read_csv("https://raw.githubusercontent.com/Luanamero/FilmesMBTI/main/Top220ImdbMbti's.csv")
```

## Introdução aos DataSets

 *   No Primeiro estão inclusas 6 colunas: Anos (Referente ao ano da premiação do Oscar 2000 - 2023), Melhor Ator (MBTI do personagem interpretado nessa categoria), Melhor Atriz (MBTI do personagem interpretado nessa categoria), Melhor Ator Quadjuvante (MBTI do personagem interpretado nessa categoria), Melhor Atriz Quadjuvante (MBTI do personagem interpretado nessa categoria), e Protagonista Melhor Filme (MBTI do personagem principal interpretado nesse filme).


*   No Segundo estão inclusas 4 colunas: NomeFilme(Nome do Filme), MBTI1(MBTI do Protagonista desse filme), MBTI2(MBTI do principal quadjuvante desse filme), MBTI3(MBTI do secundário quadjuvante desse filme).


O Primeiro e Segundo DataSets foram criados através de comparações diretas dos personagens com seus MBTI's definidos pelo Site "Personality Database" uma comunidade social orientada pelo usuário, baseada na psicologia analítica para classificação tipológica de Myers-Briggs. Ademais, eu utilizei do ranking de 250 filmes mais avaliados do IMDB para enumerar as primeiras 220 linhas do Segundo Dataset (As últimas 13 linhas incluem os filmes ganhadores de "Melhor Filme" do Oscar que não estavam inclusos nos 220).

Persoality Database: https://www.personality-database.com/

Top 250 Mais Avaliados: https://www.imdb.com/chart/top/?sort=nv,desc&mode=simple&page=1

## Análise por letra do MBTI dos Protagonistas

```python
quant_i = 0
quant_e = 0
quant_s = 0
quant_n = 0
quant_t = 0
quant_f = 0
quant_j = 0
quant_p = 0
```

```python
for mbti in dt2["MBTI1"]:
  if mbti[0] == "i":
    quant_i += 1
  elif mbti[0] == "e":
    quant_e += 1

for mbti in dt2["MBTI1"]:
  if mbti[1] == "s":
    quant_s += 1
  elif mbti[1] == "n":
    quant_n += 1

for mbti in dt2["MBTI1"]:
  if mbti[2] == "t":
    quant_t += 1
  elif mbti[2] == "f":
    #Como existe uma possibilidade de 'S' na primeira e terceira letra, coloquei s2 para diferenciá-las
    quant_f += 1

for mbti in dt2["MBTI1"]:
  if mbti[3] == "j":
    quant_j += 1
  elif mbti[3] == "p":
    #Como existe uma possibilidade de 'P' na segunda e quarta letra, coloquei p2 para diferenciá-las
    quant_p += 1

quantidade_por_elemento = [quant_i, quant_e, quant_s, quant_n, quant_t, quant_f, quant_j, quant_p]
elemento = ["i", "e", "s", "n", "t", "f", "j", "p"]

```

```python
print(f"A quantidade de Personagens Introvertidos é: {quant_i}")
print(f"A quantidade de Personagens Extrovertidos é: {quant_e}")
print(f"A quantidade de Personagens Sensitivos é: {quant_s}")
print(f"A quantidade de Personagens Intuitivos é: {quant_n}")
print(f"A quantidade de Personagens Pensativos é: {quant_t}")
print(f"A quantidade de Personagens Sentimentais é: {quant_f}")
print(f"A quantidade de Personagens 'Julgamentais' é: {quant_j}")
print(f"A quantidade de Personagens Perceptivos é: {quant_p}")
```

```python
df_elem_por_protagonista = pd.DataFrame({'MBTI': elemento, 'Quantidade': quantidade_por_elemento})

sns.set_style('whitegrid')  # Define o estilo do fundo da figura
sns.set_palette('Set2')  # Define a paleta de cores
sns.set_context('notebook', font_scale=1.2)  # Define o tamanho da fonte


sns.barplot(x='MBTI', y='Quantidade', data=df_elem_por_protagonista , color='#40E0D0', errorbar=None)
plt.title('Quantidade de Cada Elemento Entre Protagonistas', fontsize=18)
plt.xlabel('Elemento (MBTI)', fontsize=14)
plt.ylabel('Quantidade de Protagonistas', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.show()
```
